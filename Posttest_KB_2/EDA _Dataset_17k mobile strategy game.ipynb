## Analisis Dataset 1
Informasi yang didapatkan adalah: 
jumlah record, nama attribute, jumlah attribute dengan tipe objek, jumlah attribute dengan tipe angka, jumlah attribute yang memiliki nilai null.
   
df.size = digunakan untuk mengetahui jumlah seluruh elemen yang ada di dataframe

df.columns = digunakan untuk menampilkan nama-nama kolom pada dataframe

len() = Mengembalikan panjang berisi jumlah karakter dalam string atau jumlah byte yang diperlukan untuk menyimpan variabel.

df.selectdtypes = digunakan untuk mengecek tipe data untuk tiap kolom di dataframe. Bisa juga digunakan untuk mengecek tipe data salah satu kolom.

loc = digunakan untuk mengakses data berdasarkan label (nama kolom).

df.isna().any()= digunakan untuk mengecek apakah ada nilai NaN pada dataframe. 

```python
#jumlah record
df.size
```

```python
#nama atribut
for i in range (len(df.columns)):
    print(df.columns[i])
    i+=1
```

```python
#jumlah atribute dengan tipe objek
len(df.select_dtypes(['object']).columns)
```

```python
#jumlah atribut dengan tipe angka
len(df.select_dtypes(['float64','int64']).columns)
```

```python
#jumlah atribut yang memiliki nilai null
len(df.loc[:, df.isna().any()].columns)
```

## Analisi Dataset 2
Masukkan informasi di atas dalam dictionary dengan key sesuai nama informasinya. 
menggunakan variable data

```python
data={
    "jumlah record":df.size,
    "nama atribute":df.columns,
    "jumlah atribute dengan tipe object":[len(df.select_dtypes(['object']).columns)],
    "jumlah atribute dengan tipe angka":[len(df.select_dtypes(['float64','int64']).columns)],
    "jumlah atribute memiliki nilai nul":[len(df.loc[:, df.isna().any()].columns)]
    }


print("======================================================================")
print("|            Informasi pada DATASET 17K Mobile Strategy Games        |")
print("|                               Posttest 2                           |")
print("======================================================================")


print('1. Jumlah Record pada DATASET :', data.get('jumlah record'))
print("======================================================================")
print('2. Nama Atribute pada DATASET :')

for i in range (len(data["nama atribute"])):
                      print(data["nama atribute"][i])
                      i+=1
print("======================================================================")
print('jumlah atribute dengan tipe object:', data.get('jumlah atribute dengan tipe object'))
print("======================================================================")
print('jumlah atribute dengan tipe angka:', data.get('jumlah atribute dengan tipe angka'))
print("======================================================================")
print('jumlah atribute memiliki nilai nul :', data.get('jumlah atribute memiliki nilai nul'))
print("======================================================================")
```
