```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

df = pd.read_csv("17k_appstore_games.csv")
df.head(5)
```

```python
df.info()
```

# Visualisasi Dataset 1 : Grafik Batang

#### Berdasarkan Dataset dapat disimpulkan dengan Grafik batang bahwa :
Hampir kebanyakan genre pada game ditujukan untuk umur 4+, dan diantaranya ada yang hanya boleh dimainkan oleh umur 12+
dan ada sedikit dianatanya bisa dimainkan oleh umur 17+

```python
y = list(df["Age Rating"])
x = list(df["Primary Genre"])

plt.figure(figsize=(10,5))
plt.barh(x, y, color='purple')
plt.title("Rating Usia pada Genre")
plt.xlabel("Rating Usia")
plt.ylabel("Genre Utama")
plt.show()
```

# Visualisasi Dataset 2 : Grafik BoxPlot
#### Berdasarkan Dataset dapat disimpulkan dengan Grafik BoxPlot bahwa :
Rata-rata Peringkat dari pengguna berdasarkan rating umur. Rating maximum paling tinggi adalah  4+ 

```python
sns.boxplot(data = df, y = 'Average User Rating', x = 'Age Rating')
plt.show()
```

# Visualisasi Dataset 3 : Grafik Pie
#### Berdasarkan Dataset dapat disimpulkan dengan Grafik Pie bahwa :
Rata-rata Game berating tahun 4+ yaitu 69,4% , 9+ yaitu 14,6%, 12+ yaitu 12,1%, dan 17+ yaitu 3,9%


```python
genres = df.dropna(subset=['Age Rating']).sort_values(by=['Age Rating'])['Age Rating'].unique()

total = df.groupby(['Age Rating'])['Age Rating'].count()

patches, texts, pcts = plt.pie(total, labels=genres, autopct='%.1f%%', shadow = True, textprops={'size': 'x-large'})

plt.title('Rating Umur Game')
plt.setp(pcts, color='white', fontweight='bold')
plt.show()
```
